Feature Engineering

```python
df_train['Journey_day']=df_train['Date_of_Journey'].str.split('/').str[0]
df_train['Journey_month']=df_train['Date_of_Journey'].str.split('/').str[1]

df_train.Journey_day = df_train.Journey_day.astype(int)
df_train.Journey_month = df_train.Journey_month.astype(int)



#Drop the date of journey column from data set
df_train.drop(['Date_of_Journey'],axis=1,inplace=True)
```

As we extract date month and year from Date_of_Journey for better EDA

```python
df_train.head()
```

```python
# Arrival time is when the plane pulls up to the gate.
# Similar to Date_of_Journey we can extract values from Arrival_Time
#removing the date and day from arrival time column
df_train['Arrival_Time']=df_train['Arrival_Time'].str.split(' ').str[0]
#now separating the hours and minutes from Arrival_Time
df_train['Arrival_hour'] = df_train['Arrival_Time'] .str.split(':').str[0]
df_train['Arrival_min'] = df_train['Arrival_Time'] .str.split(':').str[1]


df_train.Arrival_hour = df_train.Arrival_hour.astype(int)
df_train.Arrival_min = df_train.Arrival_min.astype(int)
df_train.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df_train.head()
```

```python
# As this is case of Ordinal Categorical type we perform LabelEncoder
# Here Values are assigned with corresponding keys

df_train.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
df_train.head()
```

```python
df_train.Total_Stops.value_counts()
```

```python
df_train.Stop = df_train.Total_Stops.astype(int)
```

```python
df_train.head()
```

```python
# Departure time is when a plane leaves the gate. 
# Similar to Date_of_Journey we can extract values from Dep_Time
#same as Arrival time i am doing here
df_train['Dep_hour'] = df_train['Dep_Time'] .str.split(':').str[0]
df_train['Dep_min'] = df_train['Dep_Time'] .str.split(':').str[1]

df_train.Dep_hour = df_train.Dep_hour.astype(int)
df_train.Dep_min = df_train.Dep_min.astype(int)
df_train.drop('Dep_Time',axis=1,inplace=True)
```

```python
df_train.head()
```

```python
# Additional_Info contains almost 80% no_info
# Route and Total_Stops are related to each other

df_train.drop(["Route", "Additional_Info","Duration"], axis = 1, inplace = True)
```

```python
df_train.head()
```

```python
# As Airline is Nominal Categorical data we will perform OneHotEncoding

Airline = df_train[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first= True)

Airline.head()
```

```python
# As Source is Nominal Categorical data we will perform OneHotEncoding

Source = df_train[["Source"]]

Source = pd.get_dummies(Source, drop_first= True)

Source.head()
```

```python
# As Destination is Nominal Categorical data we will perform OneHotEncoding

Destination = df_train[["Destination"]]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
# Concatenate dataframe --> train_data + Airline + Source + Destination

df_train = pd.concat([df_train, Airline, Source, Destination], axis = 1)
```

```python
#assign the value of x and y for training and testing phase
x = df_train.loc[:,['Total_Stops','Journey_day', 'Journey_month', 'Arrival_hour',
       'Arrival_min', 'Dep_hour', 'Dep_min',
       'Airline_Air India', 'Airline_GoAir', 'Airline_IndiGo',
       'Airline_Jet Airways', 'Airline_Jet Airways Business',
       'Airline_Multiple carriers',
       'Airline_Multiple carriers Premium economy', 'Airline_SpiceJet',
       'Airline_Trujet', 'Airline_Vistara', 'Airline_Vistara Premium economy',
       'Source_Chennai', 'Source_Delhi', 'Source_Kolkata', 'Source_Mumbai',
       'Destination_Cochin', 'Destination_Delhi', 'Destination_Hyderabad',
       'Destination_Kolkata', 'Destination_New Delhi']]

x.head()
```

```python
y = df_train.iloc[:, 1]
y.head()
```

```python
#Now by using multiple Algorithms we are calculating the best Algo which suit best for our data set 

model = [DecisionTreeRegressor(),GradientBoostingRegressor(),RandomForestRegressor(),ExtraTreesRegressor()]
max_r2_score = 0
for r_state in range(40,90):
    train_x,test_x,train_y,test_y = train_test_split(x,y,random_state = r_state,test_size = 0.20)
    for i in model:
        i.fit(train_x,train_y)
        pre = i.predict(test_x)
        r2_sc = r2_score(test_y,pre)
        print("R2 score correspond to random state " ,r_state ,"is", r2_sc)
        if r2_sc> max_r2_score:
            max_r2_score=r2_sc
            final_state = r_state
            final_model = i
            
print()
print()
print()
print()
print("max R2 score correspond to random state " ,final_state , "is" , max_r2_score ,"and model is",final_model)
        
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state = 79,test_size = 0.20)
rf = RandomForestRegressor()
rf.fit(train_x,train_y)
pre = rf.predict(test_x)
print("r2 score",r2_score(test_y,pre)*100)
```

```python
print('MAE:', mean_absolute_error(test_y, pre))
print('MSE:', mean_squared_error(test_y, pre))
print('RMSE:', np.sqrt(mean_squared_error(test_y, pre)))
```

# HyperParameter Tuning

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=79, n_jobs = -1)
rf_random.fit(train_x,train_y)
```

```python
best_random_grid=rf_random.best_estimator_
pre = best_random_grid.predict(test_x)
print('R2 score',r2_score(test_y,pre)*100)
```
